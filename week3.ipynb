```python
# Name: Abubakar
# Task 1
import matplotlib.pyplot as plt

# Defining the axis values
x_values = [1, 2, 3]
y_values = [2, 4, 1]

# Ploting the line
plt.plot(x_values, y_values)

# Setting the axis labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sample graph!")

# Displaying the plot
plt.show()
```

```python
# Name: Abubakar
# Task : 2
import matplotlib.pyplot as plt

# Creating data for lines
x1= [10, 20, 30]
x2= [10,20,30]
y1 = [20, 40, 10]  # line 1 data
y2 = [40, 10, 30]  # line 2 data

# Defining tne width and colors
linewidths = [3, 5]  # line width for each line
colors = ['b', 'g']  # color for each line

# Plotting lines with labels, linewidths and colors
plt.plot(x1, y1, label='line1-width-3', linewidth=linewidths[0], color=colors[0])
plt.plot(x2, y2, label='line2-width-5', linewidth=linewidths[1], color=colors[1])

# Adding legend
plt.legend()

# Setting labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Two or more lines with different widths and colors with suitable legends')

# Show plot
plt.show()
```
